```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ratings = pd.read_csv("world-series-ratings.csv")
```

```python
plt.plot(ratings.year, ratings.average_audience)
plt.plot(ratings.year, ratings.game_1_audience)
plt.plot(ratings.year, ratings.game_2_audience)
plt.plot(ratings.year, ratings.game_3_audience)
plt.plot(ratings.year, ratings.game_4_audience)

plt.ticklabel_format(style = 'plain')
plt.xlabel("Year")
plt.ylabel("Audience (number of people)")

plt.title("World Series TV Viewership Over Time")
plt.legend(["Average Audience", "Game 1 Audience", "Game 2 Audience", "Game 3 Audience", "Game 4 Audience"], frameon = False)
```

```python
interest = pd.read_csv("mlb_interest_level.csv")
```

```python
fig, plotting = plt.subplots()

plotting.bar(interest.age, interest.avid_fan)
plotting.bar(interest.age, interest.casual_fan, bottom = interest.avid_fan)
plotting.bar(interest.age, interest.not_a_fan, bottom = np.add(interest.avid_fan,interest.casual_fan))

plt.xlabel("Age Range")
plt.ylabel("Percent of Fans")

plt.ylim(0, 150)

plt.title("MLB Fan Interest Level (As of 2022)")
plt.legend(["Avid Fan", "Casual Fan", "Not a Fan"], frameon = False)
```

```python
attendance = pd.read_csv("mlb_attendance.csv")
```

```python
plt.figure(figsize=(14, 6))
plt.bar(sorted(attendance.year), attendance.attendance, color = "maroon", width = 0.9)

plt.xlabel("Year")
plt.ylabel("Number of Fans (Millions)")

plt.title("MLB Attendance Numbers")
plt.legend(["Attendance"], frameon = False)
```
